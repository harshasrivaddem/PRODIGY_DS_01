```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/titanic.csv")
```

```python
df
```

```python
df.columns
```

```python
plt.hist(df['Age'],color="orange",edgecolor="black",alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Histogram of age distribution")
plt.show()
```

```python
sex_count = pd.value_counts(df['Sex'])
print(sex_count)
```

```python
plt.bar(sex_count.index, sex_count.values)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Bar chart of sex count")
plt.show()
```
